```python
# RECORD LINKAGE
from thefuzz import fuzz
from thefuzz import process

import pandas as pd

categories = pd.DataFrame({
    'state': ['California', 'New York', 'Texas', 'Florida', 'Ohio']
})

survey = pd.DataFrame({
    'state': ['Californa', 'New Yrok', 'Texs', 'Florda', 'Ohia', 'Texas']
})


# For each correct category
for state in categories['state']:
    # Find potential matches in states with typos
    matches = process.extract(state, survey['state'], limit=survey.shape[0])
    
    # For each potential match
    for potential_match in matches:
        # If high similarity score (>= 80)
        if potential_match[1] >= 80:
            # Replace typo with the correct category
            survey.loc[survey['state'] == potential_match[0], 'state'] = state

# Print the corrected survey DataFrame
print(survey)
```
